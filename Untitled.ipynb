```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
```

```python
df = pd.read_csv("TCPA_data_sel.csv")
df = df.drop(columns=["ID"])

chart = sns.countplot(x="Tumor", data=df)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45, horizontalalignment="right");
```

```python
ttypes= df.Tumor.unique(); ttypes[1]
```

```python
X = df[df.Tumor==ttypes[0]]
tsne = TSNE(n_components=2, perplexity=50, random_state=4, n_jobs=-1)
Xc = X.corr()
Xc[Xc < 0.3] = 0
Xemb = tsne.fit_transform(Xc)
plt.scatter(*Xemb.T)
```

```python
plt.imshow(Xc)
```
